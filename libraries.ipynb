NumPy: It, short for Numerical Python, is the core library for numerical and scientific computing in Python.

Pandas: It is a data analysis and manipulation library built on top of NumPy.

Matplotlib: It is a data visualization library that helps you create a wide variety of static, animated and interactive plots.

```python
import numpy as np
m = np.array([[1,2,3],[2,4,6],[3,5,7]])
print(m)
```
