```python
import pandas as pd
df = pd.read_csv("titanic.csv")
print(df.head())
```

```python
df.info()
```

```python
age_series = df["Age"]

print(age_series.head())
print(type(age_series))
```

```python
df["Age"].isnull().sum()
```

```python
df["Age"] = df["Age"].fillna(df["Age"].mean())
df["Sex"].value_counts()
```

```python
df["Embarked"].unique()
survived = df[df["Survived"] == 1]
print(survived.head())
```

```python
df[["Sex", "Age", "Fare"]].head()
```

```python
print(df["Age"].mean())
print(df["Fare"].max())
print(df["Fare"].min())
```

```python
plt.plot(df["Age"])
#line plot
plt.title("Age Plot")
plt.xlabel("Passenger")
plt.ylabel("Age")

plt.show()
```

```python
#histogram
plt.hist(df["Age"], bins=20)
plt.title("Age Distribution")
plt.show()
```

```python
gender_count = df["Sex"].value_counts() #bar
plt.bar(gender_count.index, gender_count.values)
plt.title("Male vs Female")
plt.show()
```

```python
survival = df["Survived"].value_counts()
plt.pie(survival.values,
        labels=["Not Survived", "Survived"],
        autopct='%1.1f%%')

plt.show()
```

Seaborn is a powerful Python library built on top of Matplotlib and Pandas that simplifies the creation of complex statistical data visualizations. It is widely used for Exploratory Data Analysis (EDA) to find patterns, trends, and relationships in data.

```python
sns.countplot(x="Pclass", data=df)#countplot
plt.show()
```

```python
sns.countplot(x = "Sex",hue="Survived", data=df)
plt.show()
```

```python
sns.boxplot(x="Pclass", y="Age", data=df)
plt.show()#box plot
```

```python
sns.scatterplot(x="Age", y="Fare", data=df)
plt.show()#scatter plot
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
plt.show()#correlation heatmap
```

```python
sns.pairplot(df)
plt.show()
```

**Univariate Analysis (One Variable)**
Univariate Analysis means analyzing only one column at a time.
Used for:
distribution
counts
mean
outliers
frequency

```python
sns.histplot(df["Age"], bins=20)
plt.title("Age Distribution")
plt.show()
```

```python
sns.countplot(x="Sex", data=df, color = 'y')
plt.title("Male vs Female")
plt.show()
```

```python
sns.boxplot(x=df["Fare"])
plt.title("Fare Boxplot")
plt.show()
```

```python
print(df["Age"].describe())
```

**Bivariate Analysis (Two Variables)**
Bivariate Analysis means analyzing relationship between two columns.
Used for:
comparison
correlation
relationship

```python
sns.countplot(x="Sex", hue="Survived", data=df)
plt.title("Survival by Gender")
plt.show()
```

```python
sns.scatterplot(x="Age", y="Fare", data=df)
plt.title("Age vs Fare")
plt.show()
```
