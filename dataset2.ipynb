```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import files
import pandas as pd

uploaded = files.upload()
df = pd.read_csv("penguins.csv")

print(df.head())
```

```python
df.info()
```

```python
print(df.isnull().sum())
```

```python
df["bill_length_mm"] = df["bill_length_mm"].fillna(df["bill_length_mm"].mean())
df["bill_depth_mm"] = df["bill_depth_mm"].fillna(df["bill_depth_mm"].mean())
df["flipper_length_mm"] = df["flipper_length_mm"].fillna(df["flipper_length_mm"].mean())
df["body_mass_g"] = df["body_mass_g"].fillna(df["body_mass_g"].mean())
```

```python
sns.countplot(x="species", data=df)
plt.title("Penguin Species Count")
plt.show()
```

```python
sns.countplot(x="island", hue="species", data=df)
plt.title("Island vs Species")
plt.show()
```

```python
sns.histplot(df["bill_length_mm"], bins=20)
plt.title("Bill Length Distribution")
plt.show()
```

```python
sns.histplot(df["body_mass_g"], bins=20)
plt.title("Body Mass Distribution")
plt.show()
```

```python
sns.scatterplot(
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="species",
    data=df
)
plt.title("Bill Length vs Bill Depth")
plt.show()
```
