```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_iris

# Load dataset
iris = load_iris()

df = pd.DataFrame(
    iris.data,
    columns=iris.feature_names
)

# Add species
df["species"] = iris.target

# Convert labels
df["species"] = df["species"].map({
    0:"setosa",
    1:"versicolor",
    2:"virginica"
})

# Basic info
print(df.head())
print(df.describe())

# Mean values using NumPy
print(np.mean(df.iloc[:,0:4], axis=0))

# Countplot
sns.countplot(x="species", data=df)
plt.show()

# Scatterplot
sns.scatterplot(
    x="sepal length (cm)",
    y="petal length (cm)",
    hue="species",
    data=df
)
plt.show()

# Heatmap
corr = df.iloc[:,0:4].corr()

sns.heatmap(corr, annot=True)

plt.show()

# Pairplot
sns.pairplot(df, hue="species")
plt.show()
```
